```python
from faker import Faker
import numpy as np
import pandas as pd
fake = Faker()
fake_names = [fake.name() for _ in range(1000)]
name_column = pd.Series(fake_names, name='name')
random_columns = np.random.rand(1000, len(df.columns) - 1)
full_df = pd.concat([name_column, pd.DataFrame(random_columns, columns=df.columns[1:])], axis=1)
full_df.head()
```
